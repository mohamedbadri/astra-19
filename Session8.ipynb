```python
#needed imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Creates an array x ranging from 0 to 1 in 100 steps
x = np.linspace(0, 1, 100)
```

```python
#Def functions for sin(x) and cos(x)
def sine_function(x):
    return np.sin(x)

def cosine_function(x):
    return np.cos(x)
```

```python
#Creates a multipanel plot with two subplots (1 row, 2 columns) and label the panels
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot sin(x) vs. x in the left panel
axs[0].plot(x, sine_function(x))
axs[0].set_title('sin(x)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')

# Plot cos(x) vs. x in the right panel
axs[1].plot(x, cosine_function(x))
axs[1].set_title('cos(x)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('cos(x')

plt.tight_layout()
plt.show()
```
