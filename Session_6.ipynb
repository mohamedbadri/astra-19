```python
# Jupyter Notebook: Sine and Cosine Table

#In this notebook, we define functions to calculate sine and cosine values and then create a table of these values versus x. We also print the first 10 values of x, sin(x), and cos(x).


### Table of Contents
#1. [Function Definitions](#cell1)
#2. [Generate Data](#cell2)
#3. [Tabulate Data](#cell3)
#4. [Print First 10 Values](#cell4)
```

```python
# Cell 1: Define functions for sin(x) and cos(x)
import math

# Define a function that returns sin(x)
def sin_x(x):
    """
    This function returns the sine of the given input x.
    """
    return math.sin(x)

# Define a function that returns cos(x)
def cos_x(x):
    """
    This function returns the cosine of the given input x.
    """
    return math.cos(x)
```

```python
# Cell 2: Generate data

# Number of data points
num_points = 1000

# Generate x values between 0 and 2
x_values = [2 * i / (num_points - 1) for i in range(num_points)]

# Calculate sin(x) and cos(x) values using the defined functions
sin_values = [sin_x(x) for x in x_values]
cos_values = [cos_x(x) for x in x_values]
```

```python
# Cell 3: Display the tabulated data

# Import the pandas library for tabulating data
import pandas as pd

# Create a DataFrame to display the data in a table format
data = {'x': x_values, 'sin(x)': sin_values, 'cos(x)': cos_values}
df = pd.DataFrame(data)

# Display the first 10 rows of the table
df.head(10)
```

```python
# Cell 4: Print the first 10 values of x, sin(x), and cos(x)

# Print a header
print("x\t\tsin(x)\t\tcos(x)")

# Print the first 10 values in columns
for i in range(10):
    print(f"{x_values[i]:.4f}\t\t{sin_values[i]:.4f}\t\t{cos_values[i]:.4f}")
```
